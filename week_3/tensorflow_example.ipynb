Let a cost function be 

J(W) = $ W^2 -10W + 25 $

This is $ (W-5)^2 $

And W will be minimized at W=5.

Let's find this minima using tensorflow framwework.

```
import numpy as np
import tensorflow as tf
```

```
# In tensorflow 2, eager execution is enabled by default
tf.executing_eagerly()
```

```
w = tf.Variable(0, dtype=tf.float32)
# `cost` is a callable that takes no argument and returns the value
# to minimize.
cost = lambda : w**2 - (10*w) + 25
# cost = lambda : tf.add(w**2, tf.multiply(-10, w), 25)

# Create an optimizer with the desired parameters.
train_opt = tf.keras.optimizers.SGD(learning_rate=0.01)

# In graph mode, returns op that minimizes the loss by updating the listed
# variables.
# op = train_opt.minimize(cost, [w])
# op.run()
# In eager mode, simply run the minimize to update the variables. It returns an operation which updates the variables in var_list
op = train_opt.minimize(cost, [w])
```

```
op
```

```
w.numpy()
```

```
# Attributes of this optimizer - No. of training iterations this optimizer has run. 
# Calling minimize automatically increments this value by 1
train_opt.iterations
```

```
# Running optimizer for 1000 iterations

for i in range(1000):
  train_opt.minimize(cost, [w])

w.numpy()
```

```
# doing it in graph mode
x = np.array([[1.], [-10], [25]])
w = tf.Variable(0, dtype=tf.float32)

cost = lambda : x[0][0]*w**2 + x[1][0]*w + x[2][0]

# def cost():
#   return x[0][0]*w**2 + x[1][0]*w + x[2][0]

train_opt = tf.keras.optimizers.SGD(learning_rate=0.01)
op = train_opt.minimize(cost, [w])
```
